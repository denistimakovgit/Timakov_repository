7. Проверить гипотезу, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом.

```python
# создадим два датафрейма для анализа. 
high_salary = df.loc[(df['salary'] == 'high')]
low_salary = df.loc[(df['salary'] == 'low')]

# применим Тест Стьюдента для сравнения двух выборок
from scipy import stats 

tvalue, pvalue = stats.ttest_ind(high_salary['average_montly_hours'], low_salary['average_montly_hours'])
print(tvalue, pvalue)

if pvalue < 0.05:
    print("""На 0.05 уровене значимости, мы отвергаем нулевые гипотезы и принимаем альтернативную.""")
    print('Сотрудники с НИЗКИМ окладом проводят на работе больше времени, чем сотрудники с низким окладом')
else:
    print("""На 0.05 уровене значимости, мы принимаем нулевые гипотезы.""")
    print('Сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом')
```

8. Рассчитать следующие показатели среди уволившихся и не уволившихся сотрудников (по отдельности)

*   Доля сотрудников с повышением за последние 5 лет
*   Средняя степень удовлетворенности
*   Среднее количество проектов





```python
# создадим два датафрейма с уволенными и не уволенными сотрудниками
retired_employees = df.loc[(df['left'] == 1)]
current_employees = df.loc[(df['left'] == 0)]
```

```python
def percent_of_num(A, B):
  'Функция для расчета процента от числа, где A - это доля от целого, а B - это общее количетво'

  per_of_num = round(((A/B)*100), 2)
  return per_of_num
```

```python
print('Доля сотрудников с повышением за последние 5 лет среди уволенных ',percent_of_num(len(retired_employees.loc[(retired_employees['promotion_last_5years'] == 1)]), len(retired_employees)),'%')
print('Доля сотрудников с повышением за последние 5 лет среди работающих ',percent_of_num(len(current_employees.loc[(current_employees['promotion_last_5years'] == 1)]), len(current_employees)),'%')
```

```python
# средняя степень удовлетворенности
print('Средняя степень удовлетворенности уволенных сотрудников ', round(retired_employees['satisfaction_level'].mean(), 2))
print('Средняя степень удовлетворенности работающих сотрудников ', round(current_employees['satisfaction_level'].mean(), 2))
```

```python
# среднее количество проектов
print('Среднее количество проектов среди уволенных сотрудников ', round(retired_employees['number_project'].mean(), 2))
print('Среднее количество проектов среди работающих сотрудников ', round(current_employees['number_project'].mean(), 2))
```
